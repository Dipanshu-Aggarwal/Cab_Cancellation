### from_area_id

```python
# from_area_id, to_area_id are categorical data - we cannot do outlier check or standarization. 

cab_data4['from_area_id'].nunique()

# there are 586 unique values in this feature. We can do binning based on the volume of bookings from areas.
# We will create a new column with volume of bookings from each area_id. 
# We will create a new label column with Low, Medium, High Booking Volume. Labelling will be done according to quantile. 
# We will also do this for cancellation colume parameter
# So we will create 4 new parameters from from_area_id. In the end we will only need only label columns. 
```

```python
# creating a column for booking volume of feature 'from_area_id'

# Adding a new coloumn at the desired position - df.insert(index, Column Heading, Column Values)
# df['Category'].map(df['Category'].value_counts())

# using value_counts we get a series whose index are the elements of 'from_area_id' and the values is the frequency count. 
# So we use map to create a series with the 'from_area_id' values replaced by the values of frequency_count.
# Finally assign this series to the 'from_area_booking_volume' column.

cab_data4.insert(1, 'from_area_booking_volume', cab_data4['from_area_id'].map(cab_data4['from_area_id'].value_counts()))
```

```python
a = 0.33
print(cab_data4['from_area_booking_volume'].quantile(0))
print(cab_data4['from_area_booking_volume'].quantile(0.33))
print(cab_data4['from_area_booking_volume'].quantile(0.67))
print(cab_data4['from_area_booking_volume'].quantile(1.00))
```

```python
# Binning the 'from_area_booking_volume' column as Low Volume, Medium Volume and High Volume. 

from_area_booking_volume_bins = [0.0, 96.0, 272.0, 3376.0]
from_area_booking_volume_bins_labels = ['Low', 'Medium', 'High']

cab_data4.insert(2, 'from_area_booking_volume_label', pd.cut(cab_data4['from_area_booking_volume'], 
                                                             bins=from_area_booking_volume_bins, 
                                                             labels=from_area_booking_volume_bins_labels))
```

```python
cab_data4.head()
```

```python
# creating a column for cancellaiton volume of feature 'from_area_id'

# The number of cancellation for a particular 'from_area_id' would be the sum of 'Car_Cancellation' column for that 'from_area_id'
# because 'Car_Cancellation' == 1 for Car Cancelled.

# Sum of values in a column = df['column_name'].sum()

from_area_car_cancellation = [] # Creating an empty list

for val in cab_data4.from_area_id:
    from_area_car_cancellation.append(cab_data4.loc[cab_data4['from_area_id']==val]['Car_Cancellation'].sum())
    # appending car_cancellation volume for each from_area_id to the list. 

# print(cab_data4.loc[cab_data4['from_area_id']==112]['Car_Cancellation'].sum())

cab_data4.insert(2, 'from_area_cancellation_volume', pd.Series(from_area_car_cancellation))
# inserting the new column with values of car_cancellation_volume
```

## ENCODING

```python
cab_data4.columns
```

```python
# function to encode fature and then remove the original feature from dataframe after it has been encoded. 

def encode_and_bind(cab_data4, feature_to_encode):
    dummies = pd.get_dummies(cab_data4[[feature_to_encode]])
    cab_data4 = pd.concat([cab_data4, dummies], axis=1)
    cab_data4 = cab_data4.drop([feature_to_encode], axis=1)
    return(cab_data4) 
```

```python
features_to_encode = ['from_area_booking_volume_label', 'from_area_cancellation_volume_label', 
                      'to_area_booking_volume_label', 'to_area_cancellation_volume_label',
                      'time_category', 'travel_day_type']
for feature in features_to_encode:
    cab_data4 = encode_and_bind(cab_data4, feature)
```

```python
fit_rf.set_params(criterion = 'gini', bootstrap = True,
                  max_features = 'auto', max_depth = 9, 
                  min_samples_split = 50, min_samples_leaf = 7)
```

```python
fit_rf.set_params(warm_start=True, 
                  oob_score=True)

min_estimators = 501
max_estimators = 1500

error_rate = {}

for i in range(min_estimators, max_estimators + 1):
    fit_rf.set_params(n_estimators = i)
    fit_rf.fit(x_train, y_train)

    oob_error = 1 - fit_rf.oob_score_
    error_rate[i] = oob_error
```

```python
# Convert dictionary to a pandas series for easy plotting 
oob_series = pd.Series(error_rate)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.set_facecolor('#fafafa')

oob_series.plot(kind='line',color = 'red')
plt.axhline(0.055, color='#875FDB',linestyle='--')
plt.axhline(0.05, color='#875FDB',linestyle='--')
plt.xlabel('n_estimators')
plt.ylabel('OOB Error Rate')
plt.title('OOB Error Rate Across various Forest sizes \n(From 501 to 1500 trees)')
```

```python
# Refine the tree via OOB Output
fit_rf.set_params(n_estimators= 200,
                  bootstrap = True,
                  warm_start=False, 
                  oob_score=False)
```
